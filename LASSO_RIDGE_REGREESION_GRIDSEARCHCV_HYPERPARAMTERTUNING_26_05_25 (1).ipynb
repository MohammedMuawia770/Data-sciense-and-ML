```python
# rEGULARIZATION Technique
# LASSO AND RIDGE
# It prevents overfitting
# adding a penalty to a loss function
# keeps the model parameters smaller and simple patterns
import pandas as pd
df=pd.read_csv(r"C:\Users\GT-499\Downloads\Advertising.csv")
```

```python
df
```

```python
from sklearn.linear_model import Lasso,Ridge
```

```python
from sklearn.metrics import r2_score
```

```python
df.shape
```

```python
df.columns
```

```python
target=df['Sales']
```

```python
target
```

```python
features=df.iloc[:,:4]
```

```python
features
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(features,target,train_size=0.75,random_state=42)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
lasso=Lasso(alpha=50)
```

```python
lasso.fit(x_train,y_train)
```

```python
lasso.fit(x_test,y_test)
```

```python
y_pred=lasso.predict(x_test)
```

```python
y_pred
```

```python
r2_score(y_test,y_pred)
```

```python
# Hyperparmeter TUning(Grid Search Cv),cross validation
params={'alpha':[10,20,30,40,50]}
# Hyperparameter tuning is a process of optimizing the configuration settings of a machine learning
# model to imorove its performance)Number of epochs:increased
```

```python
# Grid search cv is a technique used for  finding optimal parameters
from sklearn.model_selection import GridSearchCV
grid_search=GridSearchCV(lasso,params)
```

```python
grid_search.fit(x_train,y_train)
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
GridSearchCV(estimator=Lasso(alpha=50),param_grid={'alpha':[10,20,30,40,50]})
```

```python
grid_search.best_params_
```

```python
ridge=Ridge(alpha=10)
```

```python
ridge.fit(x_train,y_train)
```

```python
y_pred1=ridge.predict(x_test)
```

```python
r2_score(y_test,y_pred1)
```
