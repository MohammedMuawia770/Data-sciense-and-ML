```python
# Piechart/Pieplot :Circular graph(Categorical versus Numerical)
import matplotlib.pyplot as plt
import numpy as np
y=np.array([10,20,30,40])
x=np.array(["apple","Orange","kiwi","Banana"])
plt.pie(y,labels=x)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
y=np.array([10,20,30,50])
x=np.array(["apple","Orange","kiwi","Banana"])
z=["green","Yellow","hotpink","red"]
plt.pie(y,labels=x,colors=z)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
y=np.array([10,25,34,40])
x=np.array(["apple","Orange","kiwi","Banana"])
z=["green","Yellow","hotpink","red"]
plt.pie(y,labels=x,colors=z,autopct="%2.2f%%")
plt.show()
# autopct it is used add the percentage to your wedges"%2.3f%%""
```

```python
# explode:Function used in pyplot:To make a piechart stand out
import matplotlib.pyplot as plt
Grades=["A","B","C","D","E"]
student_marks=[40,55,100,80,25]
z=[0,0,0.2,0,0]
plt.xlabel("grades of the students")
A=["red","green","yellow","black","hotpink"]
    
plt.ylabel("Student marks")
plt.title("pieplot of grades versus students")
plt.pie(student_marks,labels=Grades,autopct="%2.1f%%",explode=z,colors=A)
plt.show()
```

```python
# How to make 2D to 3D plot
# explode:Function used in pyplot:To make a piechart stand out
# shadow=ENABLE(3D EFFECT)
import matplotlib.pyplot as plt
Grades=["A","B","C","D","E"]
student_marks=[40,55,100,80,25]
z=[0,0,0.2,0,0]
plt.xlabel("grades of the students")      
plt.ylabel("Student marks")
plt.title("pieplot of grades versus students")
plt.pie(student_marks,labels=Grades,autopct="%2.1f%%",explode=z,shadow=True)
plt.show()
```

```python
import matplotlib.pyplot as plt
Grades=["A","B","C","D","E"]
student_marks=[40,55,100,80,25]
z=[0,0,0.2,0,0.3]
plt.xlabel("grades of the students")      
plt.ylabel("Student marks")
plt.title("pieplot of grades versus students")
plt.pie(student_marks,labels=Grades,autopct="%2.1f%%",explode=z,shadow=False)
plt.show()
```

```python
# Start angle parameter is defined with an angle in degrees,default angle=0

Grades=["A","B","C","D","E"]
student_marks=[40,55,100,80,25]
z=[0,0,0.2,0,0.3]
plt.xlabel("grades of the students")      
plt.ylabel("Student marks")
plt.title("pieplot of grades versus students")
plt.pie(student_marks,labels=Grades,autopct="%2.1f%%",explode=z,shadow=False,startangle=90)
plt.show()
```
