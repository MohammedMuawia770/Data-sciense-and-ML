```python
# KMEANS IS AN UNSUPERVISED Machine Learning Algorithm
# It is used to group data points into clsuter
# similarity with each data point is used to assign a cluster
# cluster centroid are recalculated as a mean
import pandas as pd
df=pd.read_csv(r"C:\Users\GT-499\Downloads\Mall_Customers.csv")
df
#use case: we are going to analyze Annual Income and spending score using Kmeans algorithm
#step1:Import required libraries
```

```python
# step2:EDA1
# Step3:Feature Extraction

x=df.iloc[:,[3,4]].values
```

```python
x
```

```python
x.shape
```

```python
df.info()
```

```python
from sklearn.cluster import KMeans
wcss=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i,init='k-means++',random_state=42)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
# Kmeans inertia is a metric used  to evaluate how well clustering has been performed
```

```python
kmeans=KMeans(n_clusters=5,init="k-means++",random_state=42)
y_kmeans=kmeans.fit_predict(x)
y_kmeans
```

```python
import matplotlib.pyplot as plt
plt.xlabel("Annual Income")
plt.ylabel("spending score")
plt.title("Kmeans cluster")
plt.scatter(x[y_kmeans==0,0],x[y_kmeans==0,1],s=60,c="red",label="cluster1")
plt.scatter(x[y_kmeans==1,0],x[y_kmeans==1,1],s=60,c="green",label="cluster1")
plt.scatter(x[y_kmeans==2,0],x[y_kmeans==2,1],s=60,c="yellow",label="cluster1")
plt.scatter(x[y_kmeans==3,0],x[y_kmeans==3,1],s=60,c="hotpink",label="cluster1")
plt.scatter(x[y_kmeans==4,0],x[y_kmeans==4,1],s=60,c="orange",label="cluster1")
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=100,c='black')
plt.legend()
```

```python
import matplotlib.pyplot as plt
plt.xlabel("Annual Income")
plt.ylabel("spending score")
plt.title(" KMEANS CLUSTER (5 CLUSTERS)")
plt.scatter(x[y_kmeans==0,0],x[y_kmeans==0,1],s=60,c='red',label='cluster1')
plt.scatter(x[y_kmeans==1,0],x[y_kmeans==1,1],s=60,c='blue',label='cluster2')
plt.scatter(x[y_kmeans==2,0],x[y_kmeans==2,1],s=60,c='green',label='cluster3')
plt.scatter(x[y_kmeans==3,0],x[y_kmeans==3,1],s=60,c='violet',label='cluster4')
plt.scatter(x[y_kmeans==4,0],x[y_kmeans==4,1],s=60,c='yellow',label='cluster5')
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=100,c='black',label="Centroids")
plt.legend()
```

```python
label=kmeans.labels_
label
```

```python
from sklearn import metrics
print(metrics.silhouette_score(x,label))
# moderately clustering
```
