```python
import matplotlib.pyplot as plt
age_men=25,11,68,18,27,28,15,43,58,63,43,65,51,36,33,26,23,35,49,58
plt.hist(age_men,bins=10,edgecolor="black",color="red")
plt.xlabel("age of men",fontsize=10,color="green")
plt.ylabel("frequency",fontsize=20,color="black")
plt.title( "histogram plot of age of men versus frequency",fontsize=30,color="red")
plt.show()
```
