```python
# Skeweness and Kurtosis
# skeweness:It is a statistical 
# measure that describe the shape of the 
# distribution.It is a measure asymmetry of your dataset
# #skeweness=0(symmetrical)
# asymmetrical(skeweness will be positive or negative)
# # skewness=positive(right skewed,Positively skewed)
# # skewness-negative(left skewed ,Negatively skewed)
# >>>>>>>>>>>>>>>>>>
# # Kurtosis:Describes measure of peakness or flatness of
#  a distribution compared to normal distribution
# Positive Kurtosis(Leptokurtic):Indicates distribution with sharper
# peaks and heavily tailed than normal distribution
# Negative Kurtosis(Platykurtic):Indicates distribution with  flatter
# peaks and lighter tails than normal distribution
# Zero Kurtosis(Mesokurtic):Indicates a distribution with same peakness as 
# a normal distribution
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
df=pd.read_csv(r"C:\Users\GT-499\Downloads\Newspaper.csv")
```

```python
df
```

```python
sns.distplot(df['daily'])
plt.title("Daily News paper revenue")
```

```python
sns.distplot(df['sunday'])
plt.title("Daily News paper revenue")
```

```python
from scipy.stats import skew
print("skew of raw data %f"%df['sunday'].skew())
# 1.5644(skeweness)
```

```python
# df,columns
```

```python
from scipy.stats import kurtosis
print("skew of raw data %f"%df['sunday'].kurtosis())
```

```python
#Positive skeweness indicates that data distribution is longer than left tail(Rightly skewed)
# Negative skeweness indicates that data distribution is less than left tail(leftly skewed)
# skeweness=0(No skeweness)
```

```python
# In probabilty theory,kurtosis degree of peakness
# leptokurtic,heavily tailed
# Outlier in datapoint is one that signficantly difference from other observations
from scipy.stats import skew
print("skew of raw data %f"%df['daily'].skew())
```
