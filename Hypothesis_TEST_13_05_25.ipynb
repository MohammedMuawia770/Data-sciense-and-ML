```python
#One T TEST
import numpy as np
from scipy import stats
student_scores=np.array([72,89,65,73,84,83,63,76,85,75])
mu=70
#mu=population mean
```

```python
t_stat,p_value=stats.ttest_1samp(student_scores,mu)
```

```python
print("T statistics    :",t_stat)
```

```python
print("P statistics    :",p_value)
```

```python
alpha=0.05
if p_value<alpha:
    print("Reject Null Hypothesis")
else:
    print("accept NuLL Hypothesis")
```

```python
# ONE SAMPLE T TEST:(MEAN OF SINGLE SAMPLE)
# Two SAMPLE T-TEST:MEAN OF TWO SAMPLES
```

```python
import numpy as np
from scipy.stats import ttest_ind
data_group1 = np.array([14, 15, 15, 16, 13, 8, 14, 17, 16, 14, 19, 20, 21, 15, 15, 16, 16, 13, 14, 12])
data_group2 = np.array([15, 17, 14, 17, 14, 8, 12, 19, 19, 14, 17, 22, 24, 16, 13, 16, 13, 18, 15, 13])

t_stat,p_value= ttest_ind(data_group1,data_group2,equal_var=True)
```

```python
t_stat,p_value
```

```python
print("Two sample T TEST(EQUAL VAR)")
print(f"T-Statistics:{t_stat:.4f}")
print(f"P- ValueStatistics:{p_value:.5f}")
```

```python
alpha=0.05
if p_value<alpha:
    print("Reject Null Hypothesis")
else:
    print("accept NuLL Hypothesis")
```
