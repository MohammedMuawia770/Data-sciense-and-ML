```python
#Time series can leverage both supervised and unsupervised machine learning techniques
# depending on specific task and data characteristics
# Four components of Time :Trend,seasonality,irregularity and cyclicity
# aPPLICations
# Forecasting,classification,
import pandas as pd
df_temperature=pd.read_csv(r"/content/GlobalTemperatures_new.csv")
```

```python
#Simple Moving Average(SMA)
import matplotlib.pyplot as plt
df_temperature['SMA_10']=df_temperature.LandAverageTemperature.rolling(10,min_periods=1).mean()
#Simple Moving Average(SMA)
df_temperature['SMA_20']=df_temperature.LandAverageTemperature.rolling(20,min_periods=1).mean()
colors=["green","red","orange"]
df_temperature.plot(color=colors,linewidth=3,figsize=(12,6))
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.legend(labels =['LandAverageTemperature', '10-moths SMA', '20-months SMA'], fontsize=14)
plt.title('The yearly Land Average Temperature in city', fontsize=20)
plt.xlabel('Year', fontsize=16)
plt.ylabel('Temperature [°C]', fontsize=16)
```

```python
# Cumulative Moving Averag
# The  expanding() function is used to perform expanding
#window calculation on time series data within a data frame
import matplotlib.pyplot as plt
df_temperature['CMA']=df_temperature.LandAverageTemperature.expanding().mean()
colors1=['green',"Orange"]
df_temperature[['LandAverageTemperature', 'CMA']].plot(color=colors1, linewidth=3, figsize=(12,6))
plt.xticks(fontsize=8)
plt.yticks(fontsize=14)
plt.legend(labels =['Land Average Temperature', 'CMA'], fontsize=14)
plt.title('The yearly Land Average Temperature', fontsize=20)
plt.xlabel('Year', fontsize=16)
plt.ylabel('Temperature [°C]', fontsize=16)
```
