```python
# Descriptive statistics is a summary of statistics that quantitatively
#describes or summarizes features from a collection of information
#Barplot is a rectangular graph which aids to plot categorical versus
# numerical
# matplotlib is used for 2d visualization of the data
# numpy is numerical python(numerical operation)
import matplotlib.pyplot as plt
import numpy as np
x=np.array(["apple","banana","Orange","Kiwi"])
y=np.array([100,200,300,400])
plt.bar(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array(["apple","banana","Orange","Kiwi"])
y=np.array([100,200,300,400])
plt.xlabel("Fruit names")
plt.ylabel("cost of the fruits")
plt.title("bargraph of fruits versus cost of the fruits")
plt.bar(x,y)
plt.show()
```

```python
# ADD DIFFERENT COLOR AND ADD EDGES
import matplotlib.pyplot as plt
x=["villa1","villa2","villa3","villa5"]
y=[100,200,300,50]
plt.xlabel("Different villas in bangalore")
plt.ylabel("cost of villa in bangalore in dollors")
plt.title("Barplot of different villas versus cost of the villas")
plt.bar(x,y,color="red",edgecolor="black")
plt.show()
```

```python
# ADD DIFFERENT COLOR to each bar AND ADD EDGES
import matplotlib.pyplot as plt
x=["villa1","villa2","villa3","villa5"]
y=[100,200,300,50]
plt.xlabel("Different villas in bangalore")
plt.ylabel("cost of villa in bangalore in dollors")
plt.title("Barplot of different villas versus cost of the villas")
z=["red","green","yellow","hotpink"]
plt.bar(x,y,color=z,edgecolor="black")
plt.show()
```

```python
# Horizontal Plot
import matplotlib.pyplot as plt
x=("villa1","villa2","villa3","villa5")
y=(100,200,300,50)
plt.ylabel("Different villas in bangalore")
plt.xlabel("cost of villas of bangalore in dollors")
plt.title("Barplot of different villas versus cost of the villas")
z=["red","green","yellow","hotpink"]
plt.barh(x,y,color=z,edgecolor="black")
plt.show()
# There are two types of plot
# vERTICAL PLOT
# hORIZONTAL BAR PLOT
```
