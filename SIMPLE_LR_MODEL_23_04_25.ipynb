```python
sns.regplot(x=df['daily'],y=df['sunday'],data=df)
# regression plot(Linear regression plot,it is a plot of independent variable and dependent variable)
```

```python
import statsmodels.formula.api as smf
model=smf.ols('sunday~daily',data=df).fit
# # OLS:ORDINARY LEAST SQUARE ERROR METHOD
# # Formula(d1(square)+d2(square)+d3(square)+.........................dn(square))
# smf.ols(from smf import ols
```

```python
model 
```

```python
model().params
# y=beta1x+beta0 ,daily    (m)=     1.339715,beta0=1.33(y intercept),model coefficients
```

```python
model().rsquared
# ssr/sst( predicted /actual)
```
