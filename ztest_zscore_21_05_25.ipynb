```python
# ztest or Zscore is typically used when sample size is larger (More than 30 data points)
# while T test is used when sample size is smaller(Less than 30 data points)
import numpy as np
import scipy.stats as st
```

```python
sample_mean=80
population_mean=100
sample_std=15
sample_size=50
#number of samples,n=50
alpha=0.05
```

```python
z_score=(sample_mean-population_mean)/(sample_std/np.sqrt(50))
print(z_score)
```

```python
z_critical=st.norm.ppf(1-alpha)
z_critical
```

```python
if z_score > z_critical:
   print("Reject Null Hypothesis")
else:
   print("Accept Null Hypothesis")
```

```python
p_value=1-st.norm.cdf(z_score)
```

```python
p_value
```

```python
if p_value <= alpha:
   print("Reject Null Hypothesis")
else:
   print("Accept Null Hypothesis")
```
