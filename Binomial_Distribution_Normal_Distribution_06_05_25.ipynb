```python
# Binomial Distribution is a discrete distribution
#x=0,1,2,......1000
#n=number of trails
# p-probability of occurence
# size =shape of array
# EX-tossing a coin one time,probabilty of getting head /tail(0.5)- Target head
                                                                                                                                                                                                                                                                                            from numpy import random
                                                                                                                                                                                                                                                                                            random.seed(11)
                                                                                                                                                                                                                                                                                            x=random.binomial(n=10,p=0.5,size=10)
                                                                                                                                                                                                                                                                                            print(x)
```

```python
from numpy import random
random.seed(11)
x=random.binomial(n=10,p=0.5,size=100)
print(x)
```

```python
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns
sns.displot(random.binomial(n=10,p=0.5,size=1000))
```

```python
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns
sns.distplot(random.binomial(n=10,p=0.5,size=1000))
```

```python
#The normal distribution the values are unbounded
# The data points varies from minus infinity to plus infinity
# peak shaped curve(bell curve)
# loc(mean),scale(standard deviation)
from numpy import random
random.seed(20)
x=random.normal(loc=1,scale=2,size=100)
x
```

```python
from numpy import random
random.seed(20)
x=random.normal(loc=1,scale=2,size=(2,3))
x
```

```python
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns
sns.distplot(random.normal(loc=1,scale=2,size=10000))
```
