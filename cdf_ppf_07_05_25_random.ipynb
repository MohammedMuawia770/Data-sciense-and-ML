```python
# CDF:CUMULATIVE Distribution Function
# ppf:Percent Point Function
```

```python
import scipy.stats as stats
stats.uniform.cdf(x=2.5,loc=0,scale=10)
#cdf(area under curve)
```

```python
# Percent Point Function(Inverse of cdf)
stats.uniform.ppf(q=0.4,loc=0,scale=10)
# q in ppf represent the prbability for which you want to find corresponding value(quantile)
#from a given probabilty distribution
```

```python
# If you have normal distribution to find the value below which 95% of the data falls,
# you would use ppf with q=0.95.Find the ppf function is commonly used in
#statistical analysis to determin critical values
from scipy.stats import norm
q=0.95
value=norm.ppf(q)
print(value)
```
