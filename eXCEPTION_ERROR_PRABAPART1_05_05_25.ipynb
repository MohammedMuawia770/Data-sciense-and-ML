```python
# Uniform Distribution refers to a type of probability distribution
#in which all the outcomes are equally likely to occur
# The events are likely to occur if it is bounded
#The events are unlikely to occur if it is unbounded
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
uniform_data=stats.uniform.rvs(size=10000,loc=0,scale=10)
#pd.DataFrame(uniform_data).plot(kind="density",figsize=(9,9),xlim=(-1,11))
# loc=lower bound,scale=upper bound
```
